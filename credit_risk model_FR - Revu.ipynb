```python
# Cr√©ez un graphique √† barres pour visualiser les pr√™ts approuv√©s et non approuv√©s
plt.figure(figsize=(6, 6))
ax = sns.countplot(data=df, x='Pr√™tApprouv√©', color='green', width=0.4, dodge=False)  # Ajuste la largeur ici

# Ajustement des positions pour cr√©er un espace entre les barres
for i, p in enumerate(ax.patches):
    p.set_width(0.4)  # Largeur de la barre
    p.set_x(p.get_x() )  # D√©calage pour la distance

# Ajuster les ticks de l'axe x
plt.xticks([0, 1], ['Pr√™t Non Approuv√©', 'Pr√™t Approuv√©'])

# Supprimer la barre des ordonn√©es (axe y)
ax.spines['left'].set_visible(False)

# Calculer le total des pr√™ts pour le pourcentage
total = len(df)

# Afficher les pourcentages au-dessus de chaque barre
for p in ax.patches:
    height = p.get_height()
    percentage = (height / total) * 100
    ax.text(p.get_x() + p.get_width() / 2, height, f'{percentage:.1f}%', 
            ha='center', va='bottom')

# Ajouter des labels et un titre
plt.xlabel('')
plt.ylabel('')
plt.title('Pr√™ts Approuv√©s vs Non Approuv√©s')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
plt.yticks([])

# Afficher le graphique
plt.show()
```

La variable cible n'est pas √©quilibrer mais comme que ca d√©passe les 80/20 donc ce n'est pas n√©cessaire de faire un r√©√©chantillonnage. N√©anmoins, nous y tiendrons compte au moment de d√©terminer les donn√©es pour le trainning, le test et la validation.

## Step 3. Mod√©lisation

L'objectif est de construire un mod√®le pour pr√©dire les approbation des prets, dont la variable de sortie est une cat√©gorielle √† deux modalit√©s. Il s'agit donc d'une op√©ration de classification. Pour la variable de sortie, 1 repr√©sente que le pret est approuv√© et z√©ro le cas contraire.

Ainsi, nous allons utiliser 5 mod√®les de machine learning pour d√©terminer le plus performant. Entre autres nous avons le mod√®le de DecisionTreeClassifier, RandomForestClassifier, GradientBoostingClassifier, CatBoostClassifier et le XGBClassifier.

Envant de classer les donn√©es en donn√©es d'entrainement, de test et de validation. Nous allons transformer les variables cat√©gorielles en variables num√©riques.

```python
# Supprime la colonne 'Pr√™tApprouv√©' de df1 et affiche les informations de la nouvelle DataFrame df1_enc
df1_enc = df1.drop(['Pr√™tApprouv√©', 'ScoreDeRisque'], axis=1)
df1_enc.info()
```

```python
# Convertit les variables cat√©gorielles de df1_enc en variables indicatrices, en supprimant la premi√®re colonne pour √©viter la multicolin√©arit√©
X = pd.get_dummies(df1_enc, drop_first=True)
X.info()
```

```python
# La variable cible
y=df1['Pr√™tApprouv√©']
y.head()
```

```python
# S√©paration des donn√©es en ensembles d'entra√Ænement et de test
X_train, X_tr, y_train, y_tr = train_test_split(X, y, test_size=0.4, random_state=42)

# S√©paration de l'ensemble de test en ensembles de validation et de test
X_val, X_test, y_val, y_test = train_test_split(X_tr, y_tr, test_size=0.5, random_state=42)

# Affiche les dimensions des ensembles
X_train.shape, X_test.shape, X_val.shape, y_train.shape, y_test.shape, y_val.shape
```

DecisionTreeClassifier, RandomForestClassifier, GradientBoostingClassifier, CatBoostClassifier et le XGBClassifier

```python
# Initialiser les classificateurs avec une complexit√© mod√©r√©e
cls1 = CatBoostClassifier(logging_level='Silent', n_estimators=10)
cls2 = DecisionTreeClassifier()
cls3 = RandomForestClassifier(n_estimators=10)
cls4 = GradientBoostingClassifier(n_estimators=10)
cls5 = LogisticRegression(random_state=0)

# Entra√Æner les classificateurs
cls1.fit(X_train, y_train)
cls2.fit(X_train, y_train)
cls3.fit(X_train, y_train)
cls4.fit(X_train, y_train)
cls5.fit(X_train, y_train)
```

## Step 4. Resultats et Evaluation

```python
# Pr√©dire results par mod√®le
pred1 = cls1.predict(X_test)
pred2 = cls2.predict(X_test)
pred3 = cls3.predict(X_test)
pred4 = cls4.predict(X_test)
pred5 = cls5.predict(X_test)
```

```python
# Calculer l'accuracy pour chaque classificateur
acc_values = [
    accuracy_score(y_test, pred1),
    accuracy_score(y_test, pred2),
    accuracy_score(y_test, pred3),
    accuracy_score(y_test, pred4),
    accuracy_score(y_test, pred5),
]

classifier_names = [
    'CatBoostClassifier',
    'DecisionTreeClassifier',
    'RandomForestClassifier',
    'GradientBoostingClassifier',
    'LogisticRegressionClassifier', 
]

# Trier les valeurs d'accuracy par ordre d√©croissant (de haut en bas)
acc_sorted_indices = np.argsort(acc_values)[::-1]
acc_values_sorted = np.array(acc_values)[acc_sorted_indices]
classifier_names_acc_sorted = np.array(classifier_names)[acc_sorted_indices]

# Cr√©er une figure avec un seul sous-graphe
plt.figure(figsize=(9, 2))

# Tracer les valeurs d'accuracy (de haut en bas)
plt.barh(classifier_names_acc_sorted, acc_values_sorted, color='green')
plt.title('Accuracy of Classifiers (High to Low)')
plt.xlabel('Accuracy')
plt.xlim(0, 1)  # Supposant que l'accuracy varie entre 0 et 1
plt.gca().invert_yaxis()  # Meilleure performance en haut

# Afficher le graphique
plt.tight_layout()
plt.show()
```

Les mod√®les ont une tr√®s bonne performance, car chacun affiche une bonne accuracy (precision). Il semble que le Catboost pr√©sente la meilleure performance. Evaluons la courbe de auc.

```python
y_true = y_test
# Calcul des courbes ROC et AUC
fpr1, tpr1, _ = roc_curve(y_true, pred1)  
roc_auc1 = auc(fpr1, tpr1)

fpr2, tpr2, _ = roc_curve(y_true, pred2)  
roc_auc2 = auc(fpr2, tpr2)

fpr3, tpr3, _ = roc_curve(y_true, pred3)  
roc_auc3 = auc(fpr3, tpr3)

fpr4, tpr4, _ = roc_curve(y_true, pred4)  
roc_auc4 = auc(fpr4, tpr4)

fpr5, tpr5, _ = roc_curve(y_true, pred5)  
roc_auc5 = auc(fpr5, tpr5)



# Tracer les courbes ROC
plt.figure(figsize=(10, 8))

# Courbe ROC pour chaque pr√©diction
plt.plot(fpr1, tpr1, color='green', lw=2, label='ROC curve (CatBoost Classifier = {:.4f})'.format(roc_auc1))
plt.plot(fpr2, tpr2, color='blue', lw=2, label='ROC curve (DecisionTree Classifier = {:.4f})'.format(roc_auc2))
plt.plot(fpr3, tpr3, color='orange', lw=2, label='ROC curve (RandomForest Classifier = {:.4f})'.format(roc_auc3))
plt.plot(fpr4, tpr4, color='red', lw=2, label='ROC curve (GradientBoosting Classifier = {:.4f})'.format(roc_auc4))
plt.plot(fpr5, tpr5, color='indigo', lw=2, label='ROC curve (LogisticRegression Classifier = {:.4f})'.format(roc_auc5))


# Ligne diagonale
plt.plot([0, 1], [0, 1], color='black', lw=2, linestyle='--')

# Configurations des axes
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc='lower right')

# Afficher le graphique
plt.show()
```
